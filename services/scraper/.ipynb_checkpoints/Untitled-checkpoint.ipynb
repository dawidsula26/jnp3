```python
import numpy as np
import matplotlib.pyplot as plt

def generate_next_value(previous_value, minute, linear_coeff, noise_amplitude, sin_amplitude, sin_length):
    new_value = previous_value + linear_coeff
    new_value += np.random.uniform(-noise_amplitude, noise_amplitude)

    # sin_length = 10000 means that 10000 minutes = pi
    x = (minute / sin_length) % (2*np.pi)
    sin = np.sin(x)

    new_value += sin * sin_amplitude
    
    return new_value
```

```python
generate_next_value(1, 1, 0.2, 2, 1, 10)
```

```python
values = [0]

for i in range(1, 1000):
    values.append(generate_next_value(values[-1], i, 0, 20, 2, 100))
    
plt.plot(values)
```

```python
prefsum = np.cumsum(values)

plt.plot(prefsum)
```
